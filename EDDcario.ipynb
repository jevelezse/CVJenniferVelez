```python
df2 = pd.read_excel(xlsx_file,hoja_excel, header =2)
# Leer hoja específica
#df = xls.parse('NombreHoja')  # Por nombre
#df = xls.parse(0)
df = df2
```

```python
df.head()
```

```python
df1 = df[['ESTUDIO','RESULTADO']]
print('total de cariotipos:', len(df1))
```

Filtro de cariotipos normales en SP


```python
df1['RESULTADO']= df1['RESULTADO'].str.replace(' ','')
df1['RESULTADO']= df1['RESULTADO'].str.replace('\n','')
```

```python
#df1['RESULTADO']
patron = '^46,XX\[|^46,XY\[|46cromosomas'
normal = df1[df1['RESULTADO'].str.contains(patron, na=False, regex=True)]
print ('Total de cariotipos normales: ',len(normal))
```

```python
normal
```

Cariotipos con +21

```python
# Crear un patrón de expresión regular
patron1 = '\+21'

# Filtrar las filas que contengan alguno de estos patrones
resultado = df1[df1['RESULTADO'].str.contains(patron1, na=False, regex=True)]
print ('Total de cariotipos con+21: ',len(resultado))
```

```python
# Crear un patrón de expresión regular
patron2 = 'SINCRECIMIENTO'

# Filtrar las filas que contengan alguno de estos patrones
resultado2 = df1[df1['RESULTADO'].str.contains(patron2, na=False, regex=True)]
print ('Total de cariotipos cultivos negativos: ',len(resultado2))
```

```python
a = len(df)
print('Total de cariotipos anomales:',a - (len(normal) + len(resultado)))
```

```python
df_final = df1[~df1['RESULTADO'].isin(normal['RESULTADO'])]
df_final1 = df_final[~df_final['RESULTADO'].isin(resultado['RESULTADO'])]
df_final2 = df_final1[~df_final1['RESULTADO'].isin(resultado2['RESULTADO'])]
```

```python
df_final2
```

```python
patron3 = '\+'
numerico = df_final2[df_final2['RESULTADO'].str.contains(patron3, na=False, regex=True)]
print ('Total de cariotipos anormales númericos: ',len(numerico))
```

```python
anormales_e = df_final2[~df_final2['RESULTADO'].isin(numerico['RESULTADO'])]
print ('Total de cariotipos anormales estructurales: ',len(anormales_e))
```

```python
len(normal) + len(resultado) + len(resultado2) + len(numerico) +  len(anormales_e)
```

```python
print('Cariotipos normales:',len(normal)),
print('Cariotipos con +21:',len(resultado)),
print('Cariotipos cultivos negativos:',len(resultado2)),
print('Cariotipos anormales númericos:',len(numerico)),
print('Cariotipos anormales estructurales:',len(anormales_e))
```

```python
print('Cariotipos normales:',len(normal)),
print('Cariotipos con anormales:',len(resultado) + len(numerico) +  len(anormales_e))
print('Cariotipos cultivos negativos:',len(resultado2))
```
